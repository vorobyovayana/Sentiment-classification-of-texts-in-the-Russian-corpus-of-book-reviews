```python
import nltk
from nltk.collocations import *

from nltk.util import ngrams
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
with open('./bow_no_tags.txt', 'r', encoding = 'utf-8') as f:
    bow = nltk.wordpunct_tokenize(f.read())
#print(bow[:300])
with open('../swl.txt', 'r', encoding = 'utf-8')as f:
    ignored_words = f.read()

#print(ignored_words)
```

```python
finder = BigramCollocationFinder.from_words(bow)
find_trigrams = TrigramCollocationFinder.from_words(bow)
trigrams = find_trigrams.score_ngrams(trigram_measures.raw_freq)

finder.apply_freq_filter(3) # фильтрация низкочастотных коллокаций (пороговая частота 3)
finder.apply_word_filter(lambda w: w in ignored_words)
finder.nbest(bigram_measures.pmi, 100) # выделение 10 наиболее частотных биграмм, мера Pointwise Mutual Information
```

```python
finder.nbest(bigram_measures.chi_sq, 100)# выделение 10 наиболее частотных биграмм, мера Хи-квадрат
```

```python
finder.nbest(bigram_measures.likelihood_ratio, 100)# выделение 10 наиболее частотных биграмм, мера Log-Likelihood
```

```python
finder.nbest(bigram_measures.student_t, 100) # выделение 10 наиболее частотных биграмм, мера t-score
```
